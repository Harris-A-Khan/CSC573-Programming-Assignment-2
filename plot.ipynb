```python
# to plot the csv data
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read in the data files
exp1_data = pd.read_csv('experiments/sniffer_E1_renglis_hakhan2_jrskyes2_mcwise3.csv')
exp2_data = pd.read_csv('experiments/sniffer_E2_renglis_hakhan2_jrskyes2_mcwise3.csv')
exp3_data = pd.read_csv('experiments/sniffer_E3_renglis_hakhan2_jrskyes2_mcwise3.csv')
```

```python
exp1_data.head()
```

```python
# Create a figure and axis object
plt.figure(figsize=(20, 10))
ax = plt.gca()  # Get the current axis

# Plot each dataset on the same axis
exp1_data.plot(x='protocol', y='count', label='exp1', ax=ax)
exp2_data.plot(x='protocol', y='count', label='exp2', ax=ax)
exp3_data.plot(x='protocol', y='count', label='exp3', ax=ax)

# Set the x and y axis labels
ax.set_title('Packet Count by Protocol')
ax.set_xlabel('Protocol')
ax.set_ylabel('Count of Packets')

# Adding legend and showing the plot
plt.legend()
plt.show()
```

```python
# Create a figure and axis object
plt.figure(figsize=(20, 10))
ax = plt.gca()  # Get the current axis

# Width of a bar 
width = 0.2  

# Assuming 'protocol' is categorical and is the index of the DataFrame
exp1_data.plot(kind='bar', x="protocol", y='count', label='exp1', ax=ax, width=width, position=0, color='b')
exp2_data.plot(kind='bar', x="protocol", y='count', label='exp2', ax=ax, width=width, position=1, color='orange')
exp3_data.plot(kind='bar', x="protocol", y='count', label='exp3', ax=ax, width=width, position=2, color='g')

plt.ylabel('Count of Packets')
plt.xlabel('Protocol')

# Adding legend and showing the plot
plt.legend()
plt.show()
```
